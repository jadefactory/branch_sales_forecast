## 2. 데이터 전처리

### 2-1. 결측치 처리

```python
df_train = df_train.fillna(0)
df_test = df_test.fillna(0)

df_train
```

```python
df_train[['Temperature','Fuel_Price','Promotion1','Promotion2','Promotion3','Promotion4','Promotion5', 'Unemployment']].describe()
```

### 2-2. 날짜 데이터 형태 변환(일월연 -> 연월일)

```python
def arrange_date_format(date):
    year = date[6:]
    month = date[3:5]
    day = date[:2]

    return year + '-' + month + '-' + day
```

```python
df_train['Date'] = df_train['Date'].apply(arrange_date_format)
df_test['Date'] = df_test['Date'].apply(arrange_date_format)
```

```python
df_train['Date'] = pd.to_datetime(df_train['Date'])
df_test['Date'] = pd.to_datetime(df_test['Date'])

df_train.head()
```

```python
df_train.info()
```

### 2-2. 온도 데이터 형태 변환(화씨 -> 섭씨)

```python
df_train['Temperature'] = round((df_train['Temperature']-32)/1.8, 2)
df_train
```

```python
df_train['Temperature'].describe()
```

```python
df_test['Temperature'] = round((df_test['Temperature']-32)/1.8, 2)
```

## 3. 모델링

### 3-1. 모델 선택

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
train = df_train.drop(columns=['id','Date','Weekly_Sales']).copy()
target = df_train['Weekly_Sales'].copy()
```

```python
X_train, X_validate, y_train, y_validate = train_test_split(train, target, test_size=0.2, random_state=22)
```

```python
print(X_train.shape)
print(y_train.shape)
print('=============')
print(X_validate.shape)
print(y_validate.shape)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred_train = model.predict(X_train)
y_pred_validate = model.predict(X_validate)

mae_train = mean_absolute_error(y_train, y_pred_train)
print('mae_train(mae): ', mae_train)
r2_train = r2_score(y_train, y_pred_train)
print('r2_score(r2): ', r2_train)
print('')
mae_test = mean_absolute_error(y_validate, y_pred_validate)
print('mae_test(mae): ', mae_test)
r2_test = r2_score(y_validate, y_pred_validate)
print('r2_score(r2): ', r2_test)
```

```python
rfr = RandomForestRegressor(n_estimators=500, max_depth=4, random_state=1239)
rfr.fit(X_train, y_train)
```

```python
y_pred_train = rfr.predict(X_train)
y_pred_validate = rfr.predict(X_validate)

mae_train = mean_absolute_error(y_train, y_pred_train)
print('mae_train(mae): ', mae_train)
r2_train = r2_score(y_train, y_pred_train)
print('r2_score(r2): ', r2_train)
print('')
mae_test = mean_absolute_error(y_validate, y_pred_validate)
print('mae_test(mae): ', mae_test)
r2_test = r2_score(y_validate, y_pred_validate)
print('r2_score(r2): ', r2_test)
```
